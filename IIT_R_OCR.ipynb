```python
import torch
from transformers import pipeline
from transformers import AutoModel, AutoTokenizer
```

```python
from huggingface_hub import notebook_login
notebook_login()
```

```python
tokenizer = AutoTokenizer.from_pretrained('ucaslcl/GOT-OCR2_0', trust_remote_code=True)
model = AutoModel.from_pretrained('ucaslcl/GOT-OCR2_0', trust_remote_code=True, low_cpu_mem_usage=True, device_map='cuda', use_safetensors=True, pad_token_id=tokenizer.eos_token_id)
model = model.eval().cuda()
```

```python
model_id = "meta-llama/Llama-3.2-1B-Instruct"
pipe = pipeline(
    "text-generation",
    model=model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto",
)
```

```python
def ocr(image_file):
  text = model.chat(tokenizer, image_file, ocr_type='ocr')
  return text

#Quaring from the exgtracted text
def get_keywords(question):
  messages = [
      {"role": "system", "content": f"Work as an expert in finsing keywords from the any teck stack. The given text is {text}. Answer the question only from the question."},
      {"role": "user", "content": question}
  ]
  outputs = pipe(
      messages,
      max_new_tokens=1000,
  )
  return outputs[0]["generated_text"][-1]['content']
```

```python
# text in the image
image_file = '/content/llpage0001min.jpg'

text = ocr(image_file)
```
